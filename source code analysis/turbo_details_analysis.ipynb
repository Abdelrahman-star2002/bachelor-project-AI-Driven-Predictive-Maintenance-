```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.metrics import mean_squared_error, r2_score
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import random
import warnings
np.random.seed(34)
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('D:/bachelor semester/datasets/turbo_fans_with_RUL.csv')
test=pd.read_csv('D:/bachelor semester/TURBO FAN/test_data.csv')
dfrul = pd.read_csv('D:/bachelor semester/TURBO FAN/RUL_FD001.txt',header=None,index_col=False,names=['RUL'])
```

```python
from sklearn.model_selection import train_test_split
drop_labels = ['unit_number','time_cycles','setting_1','setting_2','setting_3']
train=data.drop(drop_labels,axis=1)
rul=data['RUL']
x_train, x_test, y_train, y_test=train_test_split(train,rul, test_size=0.3, random_state=42)

print('x_train:', x_train.shape, '... ', 'y_train:', y_train.shape, '... ', 'x_test:', x_test.shape, '... ', 'y_test:', y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#Droping the target variable
x_train.drop(columns=['RUL'], inplace=True)
x_test.drop(columns=['RUL'], inplace=True)
#Scaling X_train and X_test
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)

print(x_test_scale.shape)


```

```python
#linear regression
from sklearn.linear_model import LinearRegression
modellr=LinearRegression()

#svr
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')

#random_forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_features="sqrt", random_state=42)

```

```python
#evaluating functions
#R2 score & RMSE & MAER
def evaluate(y_true, y_hat, label='test'):
    mse = mean_squared_error(y_true, y_hat)
    rmse = np.sqrt(mse)
    variance = r2_score(y_true, y_hat)
    print('{} set RMSE:{}, R2:{}'.format(label, rmse, variance))


#plot real data and the predicted one to make some comparison
def plot_predActual(y_test, y_test_hat):
  
    indices = np.arange(len(y_test_hat))
    wth= 0.6
    plt.figure(figsize=(70,30))
    true_values = [int(x) for x in y_test.values]
    predicted_values = list(y_test_hat)

    plt.bar(indices, true_values, width=wth,color='b', label='True RUL')
    plt.bar([i for i in indices], predicted_values, width=0.5*wth, color='r', alpha=0.7, label='Predicted RUL')

    plt.legend(prop={'size': 40})
    plt.tick_params(labelsize=40)

    plt.show()
```

```python
#In this section we are trying to discover a time series pattern in our features 
#as we add to our data a new feature called 'rolling mean feature'. 
#It's a equivalent to the mean value of a feature considering 10 time cycles.
#It will be computed for every feature.

df=data.copy()
for x in x_train_2.columns:
    df[x+'_rm']=0
print(df.columns)

drop_labels2=['s_1', 's_5','s_6','s_10','s_16', 's_18', 's_19']
df=df.drop(columns=['setting_1','setting_2','setting_3']+drop_labels2+['RUL'], axis=1)
X_valid_3=test.drop(columns=['unit_number','time_cycles']+['setting_1','setting_2','setting_3']+drop_labels2, axis=1)

df
```

```python
#dealing with last line problem
def update_rolling_mean(data, mask):
    for x, group in mask.groupby("unit_number"):
        for x in x_train_2.columns:
            data.loc[group.index[10:], x+"_rm"] = data.loc[group.index, x].rolling(10).mean()[10:]
            data.loc[group.index[:10], x+"_rm"] = data.loc[group.index[:10], x]

update_rolling_mean(df, df)
update_rolling_mean(X_valid_3, test)

X_valid_3=X_valid_3.fillna(0)

#saving the modified dataset
df.iloc[-1,-14:]=df.iloc[-2,-14:]
X_valid_3.iloc[-1,-14:]=X_valid_3.iloc[-2,-14:]
```
